```python
# 导入哈希库
import hashlib
```

```python
# md5示例

# 初始化md5
m = hashlib.md5()

# 传入需加密的字符串,字符串需要编码成utf-8格式
m.update('使用md5加密的数据'.encode('utf-8'))

# 打印加密后的密文
print(m.hexdigest())
```

```python
# SHA256 示例

# 初始化SHA256
s = hashlib.sha256()

# 传入需要加密的字符串，同样需要编码成utf-8格式
s.update('使用SHA256加密的数据'.encode('utf-8'))

# 打印加密后的密文
print(s.hexdigest())
```

```python
# 导入base64库
import base64
```

```python
data = '你好，区块链'

# 对数据进行base64加密
result = base64.b64encode(data.encode('utf-8'))

# 打印结果
print(result)
```

```python
# 解码，base64加密是可逆的，可以密文还原
text = base64.b64decode(result)

print(text.decode('utf-8'))
```

```python
# 导入椭圆加密算法
from ecdsa import SigningKey, SECP256k1
```

```python
# 生成私钥
sk = SigningKey.generate(curve=SECP256k1)
```

```python
# 生成公钥
vk = sk.get_verifying_key()
```

```python
# 生成签名
signature = sk.sign("Something".encode("utf-8"))
```

```python
# 验证签名
vk.verify(signature, "Something".encode("utf-8"))
```
